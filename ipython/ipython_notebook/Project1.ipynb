Function definitions
--------------------

```python

def OneMax(x):
    return x.count("1")

assert OneMax("1001001") == 3

def LeadingOnes(x):
    if x and x[0] == "1": 
        return LeadingOnes(x[1:]) + 1
    return 0

assert LeadingOnes("110010011") == 2

def Jump(k):
    def Jumpk(x):
        n = len(x)
        assert k < n, "Jumpk is not defined"
        _x_ = OneMax(x)
        if _x_ < n - k:
            return _x_
        if n - k <= _x_ < n:
            return n - k
        return n
    Jumpk.__name__ = "Jump_{}".format(k)
    return Jumpk

assert Jump(2)("11000") == 2
assert Jump(2)("11110") == 3
assert Jump(2)("11111") == 5

def BinVal(x):
    return int("".join(x), 2)

assert BinVal("1100") == 12
            
def RoyalRoads(k):
    def RoyalRoadsk(x):
        n = len(x)
        assert n % k == 0, "RoyalRoadsk is not defined"
        return sum(all( char == "1" for char in x[i:i + k]) for i in range(0, n, k))
    RoyalRoadsk.__name__ = "RoyalRoads_{}".format(k)
    return RoyalRoadsk

assert RoyalRoads(3)("111000000") == 1
assert RoyalRoads(3)("110111110") == 1
```

Algorithms
----------

```python
import random

def RandomString(n):
    #return list("0") * n
    return list(bin(random.getrandbits(n))[2:].rjust(n, "0"))

assert len(RandomString(5)) == 5

def GreaterOrEqual(v1, v2):
    return v1 >= v2

def Greater(v1, v2):
    return v1 > v2

def Flip(x, i):
    x[i] = ("1" if x[i] == "0" else "0")

def RLS(f, n, stopped, better):
    x = RandomString(n)
    while not stopped(x):
        y = x[:]
        i = random.randint(0, n - 1)
        Flip(y, i)
        if better(f(y), f(x)):
            x = y
        yield x
    return x

def EA_1_plus_1(f, n, stopped, better):
    x = RandomString(n)
    while not stopped(x):
        y = x[:]
        for i in range(n):
            if random.random() < 1 / n:
                Flip(y, i)
        if better(f(y), f(x)):
            x = y
        yield x
    return x
```

```python
import collections
import time


algorithms = [RLS, EA_1_plus_1]
Jump3 = Jump(3)
RoyalRoads5 = RoyalRoads(5)
test_functions = [OneMax, LeadingOnes, Jump3, BinVal, RoyalRoads5]
comparism = [GreaterOrEqual, Greater]

def combinations():
    for better in comparism:
        for algorithm in algorithms:
            for test_function in test_functions:
                if test_function in (Jump3, RoyalRoads5) and algorithm == RLS and better == Greater:
                    continue # does not terminate
                yield better, algorithm, test_function

def steps(better, algorithm, test_function, n):
    maximum = test_function("1" * n)
    steps = 0
    #print("start", better.__name__, algorithm.__name__, test_function.__name__, n)
    for run in range(10):
        step = 0
        for step, x in enumerate(algorithm(test_function, n, lambda x: test_function(x) == maximum, better), 1):
            pass
        steps += step
    average_steps = steps / 10
    return average_steps



class X: pass

def plot(better, algorithm, test_function, n_step = 5, seconds = 5):
    n = 0
    x = []
    y = []
    name = better.__name__ + " " + algorithm.__name__ + " " + test_function.__name__
    plot = X()
    plot.x = x
    plot.y = y
    plot.name = name
    plot.n_step = n_step
    algorithm_runs[name] = plot
    while 1:
        end = time.time() + seconds
        while time.time() < end:
            n += n_step
            x.append(n)
            y.append(steps(better, algorithm, test_function, n))
        print(name)
        yield 
        
```

```python
algorithm_runs = {}
plots = []
for better, algorithm, test_function in combinations():
    plots.append(plot(better, algorithm, test_function))
print("{} plots".format(len(plots)))
```

```python
current_legend_handles = []
current_plot = None

def plot_named(name):
    global current_legend_handles, current_plot
    current_plot = plot = algorithm_runs[name]
    current_legend_handles = []
    line, = plt.plot(plot.x, plot.y, 'o', label = name)
    plt.title(name)
    plt.ylabel("steps")
    plt.xlabel("number of bits")
    current_legend_handles.append(line)
    
def plotrange():
    return range(max((1, min(current_plot.x))), max(current_plot.x) + 1)

def complexity(string, name = None):
    if name is None:
        name = string
    line, = plt.plot(plotrange(), [eval(string) for n in plotrange()], label = name)
    current_legend_handles.append(line)

def legend():
    plt.legend(handles=current_legend_handles, bbox_to_anchor=(0, 1), loc=2, borderaxespad=0.)
```
